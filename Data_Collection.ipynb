# Combine and Export CSVs

Our dataset will be made up of tech stock data exported from https://finance.yahoo.com/ from the range **FILL**. Stocks include Microsoft (MSFT), Apple (AAPL), Google (GOOG), Amazon (AMZN), Nvidia (NVDA), and Meta (META).


```python
path = ''

dfs = []
csv_files = [f for f in listdir(path) if isfile(join(path, f))]

for file in csv_files:
  # Skip the stocks.csv (combined) file if it exists
  if file == "stocks.csv": continue

  # Read csv and add it to dfs array
  df_stock = pd.read_csv(path + file, index_col=False)
  stock_name = file.replace('.csv', '')
  df_stock['stock_name'] = stock_name
  dfs.append(df_stock)

df_combined = pd.concat(dfs, axis=0)
df_combined.head()
```

```python
# Save combined stocks to Google Drive
df_combined.to_csv(path + 'stocks.csv')
```
